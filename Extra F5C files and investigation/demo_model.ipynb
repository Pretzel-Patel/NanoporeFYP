```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sample(mean, sd, p_corrupt, num, unif_min, unif_max):
    output = []
    p_corrupt = p_corrupt * np.exp(4*(np.random.rand()-0.5))
    for _ in range(num):
        if np.random.rand() > p_corrupt:
            # Sample from Gaussian
            output.append(np.random.normal(loc=mean, scale=sd))
        else:
            output.append(unif_min + np.random.rand()*(unif_max-unif_min))
            # output.append(np.random.normal(loc=100, scale=10))
    return np.array(output)
```

```python
event_means = []
event_std = []
for _ in range(3500):
    output = sample(mean=76, sd=6, p_corrupt=0.03, num=(2+np.random.geometric(0.1)), unif_min=70, unif_max=130)
    event_means.append(np.mean(output))
    event_std.append(np.std(output))

plt.scatter(event_means, event_std, marker='.', alpha=0.3);
plt.title('event_stdv vs event_level_mean using model')
plt.xlabel('event_level_mean (pA)')
plt.ylabel('event_stdv (pA)');
plt.ylim((0, 40))
plt.xlim((65, 135))
```
